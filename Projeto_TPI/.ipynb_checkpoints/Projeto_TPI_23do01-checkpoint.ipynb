### Função para ranquear a popularidade.

```python

def rangePopularidade(popularity):
    try:
        if popularity > 80:
            return 'Muito Alta'
        elif popularity > 70:
            return 'Alta'
        elif popularity > 60:
            return 'Um Pouco Alta'
        elif popularity > 50:
            return 'Moderada'
        elif popularity > 40:
            return 'Um Pouco Baixa'
        elif popularity > 30:
            return 'Baixa'
        elif popularity > 20:
            return 'Muito Baixa'
        elif popularity > 10:
            return 'Muito Muito Baixa'
        else:
            return 'Extremamente Baixa'
    except Exception as e:
        print(f'Exceção ocorrida para a popularidade {popularity}: {e}')
        return np.NaN


df['Popularidade_Range10'] = df['popularity'].apply(rangePopularidade)

def rangePopularidade4(popularity):
    if popularity > 75:
        return 'Muito Alta'
    elif popularity > 50:
        return 'Alta'
    elif popularity > 25:
        return 'Baixa'
    else:
        return 'Muito Baixa'

df['Popularidade_Range4'] = df['popularity'].apply(rangePopularidade4)
```

```python
#df.pivot_table(index='fare', values='survived', aggfunc='mean')

df.groupby('Popularidade_Range10')['duration_ms'].median()
```

```python

# Agrupar pelo intervalo de popularidade e calcular a mediana da duração
median_duration_by_popularity = df.groupby('Popularidade_Range4')['duration_ms'].median()

# Ordenar pela mediana da duração
sorted_median_duration = median_duration_by_popularity.sort_values()

# Exibindo a mediana da duração ordenada por intervalo de popularidade
print(sorted_median_duration)
```

```python

sns.boxplot(x='Popularidade_Range4', y='duration_ms', data=df)
plt.title('Distribuição da Duração por Categoria de Popularidade')
plt.show()
```

```python


# Exemplo de DataFrame
data = {'Popularidade_Range10': ['Alta', 'Baixa', 'Extremamente Baixa', 'Moderada', 'Muito Alta', 'Muito Baixa', 'Muito Muito Baixa', 'Um Pouco Alta', 'Um Pouco Baixa'],
        'explicit': [0.312608, 0.142857, 0.25, 0.252033, 0.326087, 0.1, 0.25, 0.271565, 0.232143]}

df = pd.DataFrame(data)

# Ordenar 'Popularidade_Range10'
order = ['Extremamente Baixa', 'Muito Muito Baixa', 'Muito Baixa', 'Baixa', 'Um Pouco Baixa', 'Moderada', 'Um Pouco Alta', 'Alta', 'Muito Alta']
df['Popularidade_Range10'] = pd.Categorical(df['Popularidade_Range10'], categories=order, ordered=True)
df = df.sort_values('Popularidade_Range10')

# Formatando 'explicit' como porcentagem
df['explicit'] = (df['explicit'] * 100).round(2)

# Criar a tabela pivô
pivot_table_result = df.pivot_table(index='Popularidade_Range10', values='explicit', aggfunc='mean')

# Exibindo a tabela pivô
print(pivot_table_result)
```
