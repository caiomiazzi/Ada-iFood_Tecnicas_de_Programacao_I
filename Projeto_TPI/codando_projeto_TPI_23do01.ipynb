```python
# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar o conjunto de dados
df = pd.read_csv('songs_normalize.csv')
```

```python
# Visualizar as primeiras linhas do DataFrame
#print(df.head(), end ='\n\n')

# Resumo estatístico das variáveis numéricas
#print(df.describe(), end ='\n\n')

# Contagem de valores únicos por coluna
print(df.nunique(), end ='\n\n')
```

```python
# Histograma da popularidade das músicas
plt.figure(figsize=(10, 6))
sns.histplot(df['popularity'], bins=30, kde=True)
plt.title('Distribuição da Popularidade das Músicas')
plt.xlabel('Popularidade')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de dispersão de danceability e energy
plt.figure(figsize=(10, 6))
sns.scatterplot(x='danceability', y='energy', data=df, hue='explicit', palette='Set2')
plt.title('Relação entre Danceability e Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()
```

```python
# Boxplot de valence por gênero
plt.figure(figsize=(12, 8))
sns.boxplot(x='genre', y='valence', data=df)
plt.title('Valence por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Valence')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Correlação entre variáveis numéricas
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação')
plt.show()
```

```python
# Remover colunas não numéricas antes de calcular a matriz de correlação
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Correlação entre variáveis numéricas
correlation_matrix = df_numeric.corr()

# Plotar a matriz de correlação
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação')
plt.show()
```

```python
# Selecionar apenas colunas numéricas
df_numeric = df.select_dtypes(include=['float64', 'int64'])
df_numeric.corr()
```

```python
# Selecionar apenas a linha correspondente à 'popularity' na matriz de correlação
popularity_correlation = correlation_matrix['popularity']

# Mostrar a correlação ordenada por valor absoluto
sorted_correlation = popularity_correlation.abs().sort_values(ascending=False)
print(sorted_correlation)
```

```python
# Gráfico de dispersão entre 'popularity' e 'danceability'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='loudness', y='popularity', data=df)
plt.title('Correlação entre Popularidade e Danceability')
plt.xlabel('loudness')
plt.ylabel('Popularidade')
plt.show()
```

### Relação da duração da musica com o ano de lançamento.

```python
import matplotlib.pyplot as plt

# Agrupar por ano e calcular a média da duração
avg_duration_by_year = df.groupby('year')['duration_ms'].mean()

# Plotar o gráfico de linha
plt.figure(figsize=(12, 6))
avg_duration_by_year.plot(kind='line', marker='o', linestyle='-', color='b')
plt.title('Média de Duração por Ano')
plt.xlabel('Ano')
plt.ylabel('Média de Duração (ms)')
plt.grid(True)
plt.show()
```

Verificando a quantidade de vezes que um subgenero aparece no df

```python


# Criar um DataFrame com os gêneros e suas contagens
df_genres = df['genre'].value_counts().reset_index()
df_genres.columns = ['genre', 'count']

# Criar um DataFrame para armazenar os subgêneros e suas contagens
df_subgenres = pd.DataFrame(columns=['subgenre', 'count'])

# Iterar sobre as linhas do DataFrame original
for index, row in df_genres.iterrows():
    genre = row['genre']
    count = row['count']

    # Dividir a string do gênero em subgêneros
    subgenres = [subgenre.strip() for subgenre in genre.split(',')]

    # Atualizar o DataFrame de subgêneros
    df_subgenres = pd.concat([df_subgenres, pd.DataFrame({'subgenre': subgenres, 'count': count})])

# Agrupar por subgênero e somar as contagens
df_subgenres_total = df_subgenres.groupby('subgenre')['count'].sum().reset_index()


# Ordenar por contagem total em ordem decrescente
df_subgenres_total = df_subgenres_total.sort_values(by='count', ascending=False)

# Imprimir o resultado
print(df_subgenres_total)
```

Verificando a quantidade de vezes que o termo de um genero aparece na lista de generos

```python
# Criar um DataFrame com os gêneros
df_genres = pd.DataFrame({'genre': df['genre'].unique()})

# Função para contar subgêneros
def contaSubgenero(genre_string):
    # Use str.split para dividir a string em uma lista de subgêneros
    subgeneros = [subgenre.strip() for subgenre in genre_string.split(',')]
    # Retorne a lista de subgêneros
    return subgeneros

# Aplicar a função a cada linha da coluna 'genre'
df_genres['subgenres'] = df_genres['genre'].apply(contaSubgenero)

# Criar um DataFrame para armazenar os subgêneros e suas contagens
df_subgenres = pd.DataFrame(columns=['subgenre', 'count'])

# Iterar sobre as linhas do DataFrame de gêneros
for _, row in df_genres.iterrows():
    subgenres = row['subgenres']

    # Atualizar o DataFrame de subgêneros
    df_subgenres = pd.concat([df_subgenres, pd.DataFrame({'subgenre': subgenres})])

# Contar a ocorrência de cada subgênero
subgenre_counts = df_subgenres['subgenre'].value_counts().reset_index()
subgenre_counts.columns = ['subgenre', 'count']

# Imprimir o resultado
print(subgenre_counts)
```
